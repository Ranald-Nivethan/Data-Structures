```python
class Node:
    def __init__(self, data, parent):
        self.data = data
        self.left_node = None
        self.right_node = None
        # parent may be used when implementing remove function
        self.parent = parent

class BinarySearchTree:
    def __init__(self):
        self.root = None
    #----------------------------------------------------------------------
        
    def insert_data_iteration(self,data):
        if self.root:
            node = self.root
            parent = None
            while True:
                parent = node
                if node.data<data:
                    if node.right_node:
                        node = node.right_node
                    else:
                        node.right_node = Node(data,parent)
                        break
                else:
                    if node.left_node:
                        node = node.left_node
                    else:
                        node.left_node = Node(data,parent)
                        break
        else:
            self.root = Node(data, None)

    #-----------------------------------------------------------------------

            
    def insert_data_recursion(self,data):
        if self.root:
            self.insert_data(data,self.root)
        else:
            self.root = Node(data,None)
            
    def insert_data(self,data,node):
        if data> node.data:
            if node.right_node:
                self.insert_data(data,node.right_node)
            else:
                node.right_node = Node(data, node)
        else:
            if node.left_node:
                self.insert_data(data, node.left_node)
            else:
                node.left_node = Node(data, node)
    #------------------------------------------------------------------------
                
    def find_max_min(self):
        max_val = self.maximum(self.root)
        print("max value of tree is: ",max_val)
        min_val = self.minimum(self.root)
        print("min value of tree is: ",min_val)
        
    def maximum(self,node):
        if node.right_node:
            return self.maximum(node.right_node)
        else:
            return node.data
        
    def minimum(self,node):
        if node.left_node:
            return self.minimum(node.left_node)
        else:
            return node.data
    #------------------------------------------------------------------------
                
    def inorder_traversal(self):
        self.inorder(self.root)
        print("~~THE --- END~~")
           
    def inorder(self,node):
        if node.left_node:
            self.inorder(node.left_node)
        print(node.data)
        if node.right_node:
            self.inorder(node.right_node)
    
    def preorder_traversal(self):
        self.preorder(self.root)
        print("~~THE --- END~~")
        
    def preorder(self,node):
        print(node.data)
        if node.left_node:
            self.preorder(node.left_node)

        if node.right_node:
            self.preorder(node.right_node)
            
    def postorder_traversal(self):
        self.postorder(self.root)
        print("~~THE --- END~~")
        
    def postorder(self,node):
        if node.left_node:
            self.postorder(node.left_node)

        if node.right_node:
            self.postorder(node.right_node)
        print(node.data)
    #--------------------------------------------------------
    
    def remove_element(self,data):
        node = self.find_node(self.root,data)
        print(node.data)
        prev = node.parent
        if node.left_node is None and node.right_node is None:
            if self.root == node:
                self.root.data = None
            elif prev.left_node == node:
                prev.left_node = None
            else:
                prev.right_node = None
        else:
            if node.left_node and node.right_node is None:
                self.root = node.left_node
                node = None
                
            elif node.left_node is None and node.right_node:
                self.root = node.right_node
                node = None
            else:
                max_node = self.maxi(node.left_node)
                print(max_node.data)
                if max_node.parent.left_node and max_node.parent.left_node.data == max_node.data:
                    t = max_node.data
                    max_node.data = node.data
                    node.data = t
                    max_node.parent.left_node = None
                else:
                    t = max_node.data
                    max_node.data = node.data
                    node.data = t
                    max_node.parent.right_node = None
                    
        print("element successfully removed")
                
            
    def maxi(self,node):
        if node.right_node:
            return self.maxi(node.right_node)
        else:
            return node        
            
    def find_node(self,node,data):
        
        while True:
            if node.data == data:
                break
            elif data<node.data:
                node=node.left_node
            else:
                node=node.right_node
                
        return node
            
            
        
            
            
```
