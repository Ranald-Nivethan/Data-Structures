```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class linkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.num_of_nodes = 0
        
    def add_data_front(self,data):
        node=Node(data)
        if self.head:
            node.next = self.head
            node.next.prev = node
            self.head = node
            
        else:
            self.head = node
            self.tail = self.head
              
        self.num_of_nodes+=1
    
    def add_data_back(self,data):
        node = Node(data)
        if self.head:
            self.tail.next=node
            node.prev = self.tail
            self.tail=node
        else:
            self.head = node
            self.tail = self.head
            
        self.num_of_nodes+=1
        
    def reverse_link_list(self):
        node = self.head
        prev = None
        while node.next:
            t = node.next
            node.next = prev
            prev = node
            node = t
        
        self.head = node
        self.head.next = prev
            
            
            
        
    def remove_data(self,data):
        node = self.head
        while(node.data!=data and node.next):
            node = node.next
            
        if node.data == self.head.data:
            node.next = None
        elif node.data == self.tail.data:
            node.prev = None
        elif node.data == data and node.data != self.head.data and node.data != self.tail.data:
        
            node.prev.next = node.next
            node.next.prev = node.prev
            node.next = None
            node.prev = None
            
            
        
    def print_list(self):
        print(self.head.data)
        n = self.head.next
        while(n):
            print(n.data)
            n = n.next

    def print_list_reverse(self):
        print(self.tail.data)
        n = self.tail.prev
        while(n):
            print(n.data)
            n = n.prev
        
        
        
        
        
        
    
```
